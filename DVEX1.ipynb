```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the CSV file
data = pd.read_csv('/content/5D_chart_data.csv')

# Drop the first column (Unnamed: 0, assuming it's an index column)
data = data.iloc[:, 1:]

# Set the 'Field' column as the index
data.set_index('Field', inplace=True)

# Plot the data
ax = data.plot(kind='bar', stacked=True, figsize=(10, 6))

# Annotate each part of the stacked bars with its corresponding value
for p in ax.patches:
    width, height = p.get_width(), p.get_height()
    x, y = p.get_xy()
    ax.annotate(f'{height:.1f}', (x + width/2, y + height/2), ha='center', va='center')

plt.title('Comparison of Fields Across Details')
plt.xlabel('Field')
plt.ylabel('Values')
plt.show()
```
